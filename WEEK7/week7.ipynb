```python
import pandas as pd

df = pd.read_csv('iris.data', header=None, delimiter=',')

# Show first few rows
df.head()
#cheking data types
df.dtypes

#cheking missing values
df.isnull().sum()

#checking overal information
df.info()
#statistics
df.describe()

# Group by column 0 (categorical) and compute the mean of column 1 (numerical)
grouped_means = df.groupby(0)[1].mean()

# Display result
print(grouped_means)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Preview the data
iris.head()

# Group by species and calculate mean of features
mean_values = iris.groupby('species').mean(numeric_only=True)

# Plotting
mean_values.plot(kind='line', marker='o')
plt.title('Average Feature Values by Iris Species')
plt.ylabel('Measurement (cm)')
plt.grid(True)
plt.xticks(ticks=range(len(mean_values.index)), labels=mean_values.index)
plt.legend(title='Feature')
plt.show()

# Bar chart: mean sepal length by species
sns.barplot(x='species', y='sepal_length', data=iris, ci=None)
plt.title('Average Sepal Length by Species')
plt.ylabel('Sepal Length (cm)')
plt.show()

# Scatter plot: sepal length vs sepal width
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Width')
plt.show()

# Histogram: petal length distribution
sns.histplot(data=iris, x='petal_length', kde=True, bins=20)
plt.title('Petal Length Distribution')
plt.xlabel('Petal Length (cm)')
plt.show()
```
